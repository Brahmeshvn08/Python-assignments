# ***DATA MANIPULATION AND ANALYSIS USING PANDAS***

**Overview**

*   **Why Use Pandas?**

**Data I/O and Basic Exploration**
- **Reading Data:**  
  Use `pd.read_csv()` to load data from a CSV file.
- **Previewing Data:**  
  Use `df.head()` to view the first few rows.
  
*Make sure you have the file* `sample_data.csv` *in your working directory.*

```python
#Importing of pandas library
import pandas as pd
pd.read_csv('/content/electrical_consumption.csv')
```

```python
df=pd.read_csv('/content/electrical_consumption.csv') #Assigning Dataset to a variable df
df.head() #Displays the first few rows of the DataFrame
```

```python
df.tail() #Displays the last few rows of the DataFrame
```

```python
df.head(300) #views first 300 rows
```

```python
df.tail(300) #views last 300 rows
```

```python
df2=df.head(300) #Assigning 300 rows to a new Dataframe 2 (df2) variable
df2.head() #Displays the first few rows of Dataframe 2
```

```python
df2.info() #Gets a concise summary of the DataFrame
```

```python
df2.describe() #Generates descriptive statistics
```

```python
df2.shape #Gets the dimensions of the DataFrame
```

```python
df2.columns #Lists the column names
```

```python
df2.dtypes #Display data types of each column
```

```python
df2['Meter_Status'] #Selects a single column
```

```python
df2[['Consumption_kWh','Billing_Amount','Meter_Status']] #Selects multiple columns
```

```python
df2.iloc[3:9] #Selects data by integer-based indexing
```

```python
df2.loc[3:9,'Consumption_kWh':'Power_Factor'] #Selects data by label-based indexing
```

```python
df2[df2['Billing_Amount']>70] #Filters rows based on a condition
```

```python
df2.groupby('Customer_ID')[['Consumption_kWh','Billing_Amount',]].mean() #Groups data for aggregation operations
```

```python
df2.isnull().sum() #Detects the total no.of missing values from each column
```

```python
df2.dropna() #Remove rows with missing values
```

```python
df2.duplicated() #Checks for duplicate rows
```

```python
df2.sort_values(['Billing_Amount','Power_Factor']) #Sort the DataFrame by one or more columns
```

```python
df2['Power_Factor'].fillna(0.96,inplace=True) #Fill missing values with a specified value
```

```python
df2.head()
```

```python
df2['Load_Factor']=round(df2['Consumption_kWh']/df2['Peak_Hours_Consumption'],2) #Add a new column to the DataFrame
```

```python
df2.head()
```

```python
df2=df2.rename(columns={'Consumption_kWh':'Total_Consumption_kWh'}) #Rename DataFrame columns
```

```python
df2.head()
```

```python
df2=df2.set_index('Customer_ID') #Set a column as the index
```

```python
df2.head()
```

```python
df2.reset_index() #Reset the DataFrame index
```

```python
df2.to_csv() #Write the DataFrame to a CSV file
```

# ***MATPLOTLIB***

**Overview**
- **Purpose:**  
  Matplotlib is the go-to library for creating visualizations in Python.
- **Flexibility:**  
  Offers extensive customization options for plot styling, axes, labels, legends, and more.

**Basic Plotting**

  Line Plot: Salary vs. Age

*   Line Plot: Salary vs. Age
*   Visualize the relationship between Age and Salary using a line plot.

```python
df2=df2.head(5)
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt

# Line Plot: Peak_Hours_Consumption vs. Total_Consumption_kWh using the DataFrame (df)
plt.figure(figsize=(6, 4))
plt.plot(df2['Peak_Hours_Consumption'], df2['Total_Consumption_kWh'], marker='o', linestyle='-')
plt.title('Peak_Hours_Consumption vs. Total_Consumption_kWh')
plt.xlabel('Peak_Hours_Consumption')
plt.ylabel('Total_Consumption_kWh')
plt.grid(True)
plt.show()
```

#### Bar Chart:
Use a bar chart to visualize aggregated data.

```python
# Bar Chart: Average of Peak_Hours_Consumption across different meter statuses.
df2.plot(kind='bar')
```
